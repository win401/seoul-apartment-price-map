## STEP 1: 사전 준비

```python
from openai import OpenAI
import json

# API 키 설정
try:
    from google.colab import userdata
    api_key = userdata.get('OPENAI_API_KEY')
except:
    api_key = input("OpenAI API 키를 입력하세요: ")

client = OpenAI(api_key=api_key)

def call_gpt(prompt, model="gpt-4o-mini"):
    """GPT API를 호출하고 응답을 반환하는 함수"""
    response = client.chat.completions.create(
        model=model,
        max_tokens=1000,
        temperature=0,
        messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content

print("✅ 준비 완료!")
```

## STEP 2: 샘플 제품 리뷰 데이터

```python
# 샘플 리뷰 데이터
reviews = [
    """제품명: 스마트워치 X100
리뷰: 구매한지 2개월 되었는데, 배터리가 너무 빨리 떨어져요.
처음 1주일은 괜찮았지만 지금은 하루도 못가요.
디자인은 마음에 드는데 기능이 아쉬워요.
배터리 개선되면 별 5개 줄텐데...""",

    """제품명: 스마트워치 X100
리뷰: 정말 만족해요!
가볍고 세련된 디자인에 운동 추적 기능이 정말 정확합니다.
건강 관리하는데 많은 도움이 됩니다.
이 가격에 이 정도면 가성비 최고예요.""",

    """제품명: 스마트워치 X100
리뷰: 화면이 너무 작아서 글씨가 잘 안 보여요.
시간 확인할 때도 정신을 집중해야 합니다.
화면 크기만 좀 더 크면 좋겠어요.
그 외는 무난해요."""
]

print("샘플 리뷰 데이터 로드 완료!")
print(f"총 {len(reviews)}개의 리뷰")
```

## STEP 3: 감정 분석 + 주요 이슈 추출

```python
def analyze_review(review):
    """리뷰를 분석하고 구조화된 결과 반환"""
    prompt = f"""다음 제품 리뷰를 분석해주세요.
감정: (긍정/부정/중립)
별점: (1~5)
주요 장점:
주요 단점:

리뷰:
{review}
"""
    return call_gpt(prompt)

print("\n리뷰 분석 시작...")
for i, review in enumerate(reviews, 1):
    print(f"\n리뷰 {i}:")
    result = analyze_review(review)
    print(result)
```
